```python
import pandas as pd
import pandas_datareader.data as wb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from datetime import datetime
start = datetime(1970,1,1)
end = datetime(2016,7,20)

df = wb.DataReader("^N225","yahoo",start,end)
```

```python
data = df["Adj Close"]
print len(data)
data = list(data)
```

```python
def split_data(data):
    
    train,val,test=data[:5125],data[5125:(5125+1285)],data[(5125+1285):]
    
    return train,val,test

train,val,test=split_data(data)
print len(train),len(val),len(test)
print train
```

```python
def group_data(data):
    data_ = []
    for t in xrange(len(data)-25):
        data_.append(data[t:t+25:5])
    return data_
train = group_data(train)
val = group_data(val)
test = group_data(test)
print len(train),len(val),len(test)
```

```python
def split_xy(data):
    data_x=[]
    data_y=[]
    for i in xrange(len(data)-25):
        k = data[i:i+25:5]
        data_x.append(k)
        j = data[5+i]
        data_y.append(j)
    data_x = np.array(data_x,dtype=np.float32)
    data_y = np.array(data_y,dtype=np.float32)
    return data_x,data_y
train_x ,train_y = split_xy(train)
val_x,val_y = split_xy(val)
test_x,test_y = split_xy(test)
```

```python
def organize_data(data):
    data_ = np.empty(5)
    for i in data:
        x = np.array(i)
        data_ = np.vstack((data_,x))
    data_ = data_[1:]
    return data_
train = organize_data(train)
val = organize_data(val)
test = organize_data(test)
```

```python
def label_data(data,time_steps,labels=False):
    df = []
    for i in xrange(len(data)-time_steps):
        if labels:
            try:
                df.append(data[i+time_steps]).as_matrix()
            except AttributeError:
                df.append(data[i+time_steps])
        else:
            data_ = data[i:i+time_steps]
            df.append(data_ if len(data_.shape) > 1 else [[i] for i in data_])
    return np.array(df)
time_steps = 5
train_x = label_data(train_np,time_steps,labels=False)
train_y = label_data(train_np,time_steps,labels=True)
```
